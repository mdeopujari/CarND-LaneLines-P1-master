## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

#global L_slopes = [-0.6] # dummy value in array that holds upto 5 recent slopes together
#global R_slopes = [0.6]  # dummy value in array that holds upto 5 recent slopes together

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[128, 128, 128], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    rx = [] # set of x-coordinates for right line
    ry = [] # set of y-coordinates for right line
    lx = [] # set of x-coordinates for left line
    ly = [] # set of y-coordinates for left line
           
    for line in lines:
        for x1,y1,x2,y2 in line:
            if((x1!=x2) and (y1!=y2)):
                slope = (y2-y1)/(x2-x1) # calculating slope
                
                if (slope>=0): # if positive then its the right side line
                    rx.append(x1)
                    rx.append(x2)
                    ry.append(y1)
                    ry.append(y2)
       
                else:          # else it is the left side line  
                    lx.append(x1)
                    lx.append(x2)
                    ly.append(y1)
                    ly.append(y2)
                    
            #cv2.line(img, (x1, y1), (x2, y2), color, thickness)
      
    # Draw Right Lane
    if(rx and ry): # when both sets are not empty
        fit_right = np.polyfit(rx,ry,1)                   # Use polyfit for regression
        
        if(fit_right[0] < 0.5): # thresholding the slope
            fit_right[0] = 0.5
        if(fit_right[0] > 0.8):
            fit_right[0] = 0.8            
        
        RY1 = imshape[0]                                  # Bottom of image Y-coordinate
        RX1 = int( (RY1 - fit_right[1]) / fit_right[0] )  # Bottom of image X-coordinate
        
        RY2 = int (imshape[0]/2 + 50)                     # Top of ROI Y-coordinate
        RX2 = int( (RY2 - fit_right[1]) / fit_right[0] )  # Top of ROI X-coordinate
        
        cv2.line(img, (RX1,RY1), (RX2,RY2), [0,0,255], 10) # Annotate Right-lane
        
       # if(fit_right[0] > 0.5) and fit_right[0] < 0.8):   # threshold the slope
       #     cv2.line(img, (RX1,RY1), (RX2,RY2), [0,0,255], 10) # Annotate Right-lane     
         
    # Draw Left Lane
    if(lx and ly): # when both sets are not empty
        fit_left = np.polyfit(lx,ly,1)                 # Use polyfit for regression           
        
        if(fit_left[0] > -0.5): # thresholding the slope
            fit_left[0] = -0.5
        if(fit_left[0] < -0.8):
            fit_left[0] = -0.8 
        
        LY1 = imshape[0]                               # Bottom of image Y-coordinate  
        LX1 = int( (LY1 - fit_left[1]) / fit_left[0] ) # Bottom of image X-coordinate 
        
        LY2 = int(imshape[0]/2 + 50)                   # Top of ROI Y-coordinate
        LX2 = int( (LY2 - fit_left[1]) / fit_left[0] ) # Top of ROI X-coordinate        
        
        cv2.line(img, (LX1,LY1), (LX2,LY2), [0,0,255], 10) # Annotate Left-lane
        
       # if(fit_left[0] > -0.8 and fit_left[0] < -0.5): # threshold the slope
       #     cv2.line(img, (LX1,LY1), (LX2,LY2), [0,0,255], 10) # Annotate Left-lane
                

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.

for filename in os.listdir("test_images/"):
    ## Read in image
    image = mpimg.imread("test_images/" + filename)
    
    ## Copying original image
    raw_image = np.copy(image)  
    color_filtered_image = np.copy(image)
    
    ## Define color selection criteria    
    
    # Highlighting white stuff, Filtering/Masking out non-white stuff 
    white_image = cv2.inRange(raw_image,np.array([210,210,210]), np.array([255,255,255]))    
        
    # Highlighting yellow stuff, Filtering/Masking out non-yellow stuff    
    hsv_image = cv2.cvtColor(raw_image,cv2.COLOR_RGB2HSV)
    yellow_image = cv2.inRange(hsv_image,np.array([9,100,100]), np.array([29,255,255]))
        
    color_filtered_image = yellow_image | white_image    
    
    # image is already grayscale
    gray_image = color_filtered_image     
    
    ## Apply smoothing over grayscale image
    KERNEL_SIZE = 3
    smoothed_image = gaussian_blur(gray_image,KERNEL_SIZE)
    
    ## Apply Canny Edge Detection
    low_threshold = 50
    high_threshold = 150
    edges_image = cv2.Canny(smoothed_image, low_threshold, high_threshold)
    
    ## Define vertices for Region of Interest
    imshape = image.shape    
    vertices = np.array([[(60, imshape[0]),(imshape[1]/2 - 40,imshape[0]/2 + 50), (imshape[1]/2 + 40,imshape[0]/2 + 50), (imshape[1]-60,imshape[0])]], dtype=np.int32)
    
    ## Obtain masked Region of Interest image where only ROI is filled with fillcolor (255)
    masked_roi = region_of_interest(edges_image, vertices)    
    
    ## Define the Hough transform parameters 
    # https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_imgproc/py_houghlines/py_houghlines.html
    rho = 1              # distance resolution in pixels of the Hough grid
    theta = np.pi/180    # angular resolution in radians of the Hough grid
    threshold = 15       # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 # minimum number of pixels making up a line
    max_line_gap = 5     # maximum gap in pixels between connectable line segments
    
    ## Run Hough on edge detected image
    line_image = hough_lines(masked_roi, rho, theta, threshold, min_line_length, max_line_gap)           
    
    overlayed_image = weighted_img(line_image, raw_image, α=0.8, β=1., γ=0.)
    
    mpimg.imsave("test_images_output/" + filename, overlayed_image)    
    plt.imshow (overlayed_image)
````

````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)    
    
    ## Copying original image
    raw_image = np.copy(image)
    color_filtered_image = np.copy(image)
    
    ## STEP 1 - Color Selection
    
    # STEP 1a) - Highlighting white stuff, Filtering/Masking out non-white stuff 
    white_image = cv2.inRange(raw_image,np.array([210,210,210]),np.array([255,255,255]))    
    
    # STEP 1b) - Highlighting yellow stuff, Filtering/Masking out non-yellow stuff    
    hsv_image = cv2.cvtColor(raw_image,cv2.COLOR_RGB2HSV) # Converting to HSV space for yellow color
    yellow_image = cv2.inRange(hsv_image,np.array([9,100,100]),np.array([29,255,255]))
    
    color_filtered_image = yellow_image | white_image   # This is already a Grayscale image    
    
    ## STEP 2 - Edge Detection
    
    # STEP 2a) - Apply smoothing over grayscale image
    KERNEL_SIZE = 3
    smoothed_image = gaussian_blur(color_filtered_image,KERNEL_SIZE)
    
    # STEP 2b) - Apply Canny Edge Detection 
    low_threshold = 50
    high_threshold = 150
    edges_image = cv2.Canny(smoothed_image, low_threshold, high_threshold)
    
    ## STEP 3 - Region of Interest
    
    # STEP 3a) -  Define vertices for Region of Interest
    imshape = image.shape
    V1 = (60, imshape[0])                         # Bottom-left vertex of ROI
    V2 = (imshape[1]/2 - 40, imshape[0]/2 + 50)   # Top-left vertex of ROI
    V3 = (imshape[1]/2 + 40, imshape[0]/2 + 50)   # Top-right vertex of ROI
    V4 = (imshape[1] - 60, imshape[0])            # Bottom-right vertex of ROI
    
    vertices = np.array([[V1,V2,V3,V4]], dtype=np.int32)
    
    # STEP 3b) -  Obtain masked Region of Interest image where only ROI is filled with fillcolor (255)
    masked_roi = region_of_interest(edges_image, vertices)    
    
    ## STEP 4 - Hough Transform to detect lane lines
    
    # STEP 4a) -  Define the Hough transform parameters    
    rho = 1              # distance resolution in pixels of the Hough grid
    theta = np.pi/180    # angular resolution in radians of the Hough grid
    threshold = 15       # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 20 # minimum number of pixels making up a line
    max_line_gap = 2     # maximum gap in pixels between connectable line segments
    
    # STEP 4b) -  Run Hough on edge detected image
    line_image = hough_lines(masked_roi, rho, theta, threshold, min_line_length, max_line_gap)        
       
    ## STEP 5 - Overlay the detected lane line on original image    
    overlayed_image = weighted_img(line_image, raw_image, α=0.8, β=1., γ=0.)    

    return overlayed_image
````
